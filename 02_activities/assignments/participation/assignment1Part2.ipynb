```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn


np.random.seed(613)
x= np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)
```

```python
# plot both y variables on the same axis
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
#adding a legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )# - label: The name of this line, to appear in the legend
ax.plot(x,y2, label = "Person 2" ) # - label: The name of this line, to appear in the legend
ax.legend(loc='lower right')  #Add a legend to the plot
fig.show()
```

```python
#Modify legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")

#Add a legend to the plot with customization options

ax.legend(loc='lower right',    #Specifies the position of the legend in the lower-right corner
        frameon = True,         #add frame around the legend for better visibility
        fontsize = 12,           #Sets the font size of the text inside the legend
        ncol = 2,                #Arranges the legend entries in 2 columns
        shadow = True )          # Adds a shadow effect to the legend box for aesthetics
fig.show()
```

```python
#Moving legend outside of the plot area

fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")

ax.legend(loc= "upper left",                 # Places the legend's anchor point in the upper-left corner of the plot area
        bbox_to_anchor =(1, 1))                # Adjusts the legend's position to be outside the plot

fig.show()
```

```python
#Annotations with text() function
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.text(10, 95, "This value is important!") # Add text at coordinates (10, 95)
fig.show()
```

```python
#Modify text annotations
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!",
      ha='center',    #alignment of text
      color = 'red',  #modify font colour
      size = 20)      #modify font size
fig.show()
```

```python
#Positioning text annotations
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10]) #Defines the range for x-axis (0 to 10) and y-axis (0 to 10)

#Add text using data coordinates
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData    ) # Add text at position (1, 5) using data coordinates; transform=ax.transData: Positions the text based on the data values

#Add text using axes coordinates
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes) # (0.5, 0.1): Position of the text in **axes-relative coordinates**, transform=ax.transAxes: Interprets coordinates as fractions of the axes
#   (0.5 = 50% of the width, 0.1 = 10% of the height of the axes)

#Add text using figure coordinates
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)",transform=fig.transFigure )#(0.2, 0.2): Position of the text in **figure-relative coordinates**transform=fig.transFigure: Interprets coordinates as fractions of the entire figure
#   (0.2 = 20% of the figure width, 0.2 = 20% of the figure height)
```

```python
#annotating with arrows
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),  #xy=(10, 95) : point on the plot where the annotation arrow points to, xytext=(20, 94):position where the text will be displayed, arrowprops --> Customizes the appearance of the arrow
          arrowprops=dict(facecolor='black', shrink=0.05)) #facecolor='black': The color of the arrow ,shrink --> Shrinks the size of the arrow to make it smalle
fig.show()
```

Matplotlib annotations documentation :   https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.annotate.html

```python
#Removing tick marks and labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.yaxis.set_major_locator(plt.NullLocator()) #Remove the major ticks from the y-axis , plt.NullLocator() removes the tick marks and labels from the y-axis
ax.xaxis.set_major_formatter(plt.NullFormatter()) # Remove the major ticks and labels from the x-axis, plt.NullFormatter() removes the tick labels from the x-axis
```

```python
#limiting tick marks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #Set the x-axis to display a maximum of 3 ticks, plt.MaxNLocator(3) automatically places at most 3 ticks on the x-axis
```

```python
#Tick mark intervals using MultipleLocator
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))  #places ticks on the x-axis at every multiple of 5 units
```

```python
#Rotating axis labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xticks(rotation=45, ha='right') # Rotate x-tick labels by 45 degrees and align them to the right

```

```python
#Modifying axis title fonts 
font1 = {'family':'serif','color':'indigo'}

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1)
```
