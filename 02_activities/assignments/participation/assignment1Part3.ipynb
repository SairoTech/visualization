**Subplots**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x1 = np.arange(50) #generates a sequence of integers from 0 to 49
y1 = np.random.randint(0, 75,50) #generates an array of 50 random integers, each between 0 (inclusive) and 75 (exclusive)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"]) #creates a NumPy array x2 with 5 elements
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# create two plots ax1 and ax2 - creates a figure with two subplots arranged side-by-side, each displaying different types of plots. 
fig,(ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
```

```python
#Activity: Customizing our plots 
#Create subplots
fig,(ax1, ax2) = plt.subplots(ncols=2,
                        nrows=1,
                        figsize=(7, 3))

# Customizing the scatter plot (ax1)
ax1.scatter(x1,y1,color='green', marker='o', label='Scatter Points')
ax1.set_title('Scatter Plot', fontsize=14, fontweight='bold', family='Arial')
ax1.set_xlabel('X Values', fontsize=12)
ax1.set_ylabel('Y Values', fontsize=12)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend(fontsize=10)
#ax1.annotate('Peak Point', xy=(x1[5], y1[5]), xytext=(x1[5]+1, y1[5]+10),
             #arrowprops=dict(facecolor='red', arrowstyle="->"),
             #fontsize=10, color='blue')

#customizing bar plot(ax2)
ax2.bar(x2, y2, color='orange', label='Bar Data')  # Change bar color
ax2.set_title('Bar Plot', fontsize=14, fontweight='bold', family='Times New Roman')  # Title with font settings
ax2.set_xlabel('Categories', fontsize=12)
ax2.set_ylabel('Values', fontsize=12)
ax2.grid(True, axis='y', linestyle=':', color='black')  # Y-axis grid lines

# Add annotation in bar plot
#ax2.annotate('Max Value', xy=(2, y2[2]), xytext=(3, y2[2]+20),
             #arrowprops=dict(facecolor='blue', arrowstyle="->"),
             #fontsize=10, color='purple')
fig.tight_layout(pad=3.0)  # without this line, the two plots were overlapping
fig.show()
```

```python
#Subplots without a grid arrangement
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
```

```python
#replicated scenario to understand when to use layouts
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel("A Big Label", fontsize = 18)
someaxes["ax2"].set_xlabel("Another Label", fontsize = 18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize = 18)
```

```python
#use layout 
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                ['ax2', 'ax3']],
                                figsize=(7, 4),
                                layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

someaxes["ax1"].set_xlabel("A Big Label", fontsize = 18)
someaxes["ax2"].set_xlabel("Another Label", fontsize = 18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize = 18)
```
