```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn
```

```python
#set seed
np.random.seed(613)

#generate data
x = np.arange(50) #0,1,2,3,4,.....49
y = np.random.randint(0, 100, 50) # first value is our min, second value is our max, third value is how many to generate/number of observations

#scatter plot
fig, ax = plt.subplots(figsize=(5, 3)) # creates a figure 'fig' and an axis 'ax' for plotting and 'figsize' parameter specifies
                                       # the size of the figure in inches, where (5,3) means 5 inches wide and 3 inches tall
ax.scatter(x, y)
```

subplots() : function in matplotlib to create a figure and one or more axes in a single step. fig is the canvas where the plots are drawn. it is like a big container for all the subplots, titles and labels.

create axes : the ax object represents the actual plotting area inside the figure. it is where you plot data like lines, histograms, scatter plots etc.

```python
#bar plot
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
#line plot
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
#Histogram
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y) # see my notes below
```

ax.hist(y) : A histogram doesnt need x because its not about plotting points. Instead, it groups the values in y into ranges (called bins) and counts how many values fall into each bin.The x-axis in a histogram is automatically created based on the bins (ranges of numbers, like 010, 1020, etc.).

Plot labels

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time') #title
ax.set_ylabel('Total growth') #Y axis label
ax.set_xlabel('Years since start') # X axis label
```

```python
#Setting fonts
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color': 'green', 'size':14}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)


ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
#adjusting dot type
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y,
            marker = '*', # 'o'/'+'
            color = 'indigo')
```

```python
#line style
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
            marker = '*',
            color = 'indigo',
            linestyle = '--',
            linewidth = 2)
```

```python

#marker color
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
            marker = '*',
            color = 'indigo', #line color
            linestyle = '--',
            linewidth = 2,
            markeredgecolor = '#FA9359', #marker edge color
            markerfacecolor = '#000000') #marker inside color

ax.grid() # both vertical and horizontal grid lines

#ax.grid(axis = "x") # vertical grid lines

#ax.grid(axis = "y") # horizontal grid lines
```

```python

#Activity to modify appearance of graph 

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y,
            marker = '*',
            color = 'gray',
            linestyle = '--',
            linewidth = 2)
ax.grid(linestyle='-.', linewidth = 1.5) # both vertical and horizontal grid lines
```
